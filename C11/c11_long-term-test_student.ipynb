```python
import torch
import torch.nn as nn
import numpy as np
```

#### CZĘŚĆ 1: GENEROWANIE TRUDNYCH DANYCH (Nie modyfikuj)

```python
# Tworzymy 1000 sekwencji. Każda ma 50 kroków czasowych, 
# ale tylko 1 cechę (liczbę) w każdym kroku.
wielkosc_zbioru = 1000
dlugosc_sekwencji = 50
rozmiar_wejscia = 1

X_dane = np.random.rand(wielkosc_zbioru, dlugosc_sekwencji, rozmiar_wejscia)
Y_dane = np.zeros(wielkosc_zbioru)
```

```python
# ZASADA GRY: O klasie (0 lub 1) decyduje TYLKO pierwsza liczba w sekwencji!
for i in range(wielkosc_zbioru):
    if np.random.rand() > 0.5:
        X_dane[i, 0, 0] = 1.0  # Ustawiamy sygnał na 1.0 na samym początku
        Y_dane[i] = 1          # Klasa 1
    else:
        X_dane[i, 0, 0] = 0.0  # Ustawiamy sygnał na 0.0 na samym początku
        Y_dane[i] = 0          # Klasa 0
        
X = torch.tensor(X_dane, dtype=torch.float32)
Y = torch.tensor(Y_dane, dtype=torch.long)

print(f"Kształt wejścia X: {X.shape} -> [Batch, Czas, Cechy]")
```

#### CZĘŚĆ 3: TRENING (Nie modyfikuj)

```python
# Inicjalizacja
model = DetektywLSTM(rozmiar_wejscia=1, rozmiar_pamieci=16, liczba_klas=2)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

print("\nRozpoczynam badanie pamięci...")
epoki = 150

for epoka in range(epoki):
    optimizer.zero_grad()
    wyjscie = model(X)
    blad = criterion(wyjscie, Y)
    
    blad.backward()
    optimizer.step()
    
    # Obliczanie dokładności (Accuracy)
    _, przewidywania = torch.max(wyjscie, 1)
    poprawne = (przewidywania == Y).sum().item()
    dokladnosc = poprawne / wielkosc_zbioru * 100
    
    if (epoka+1) % 30 == 0:
        print(f"Epoka: {epoka+1}/{epoki} | Błąd: {blad.item():.4f} | Dokładność: {dokladnosc:.1f}%")

print("\nZadanie Dodatkowe dla ambitnych:")
print("Zmień w swoim kodzie nn.LSTM z powrotem na nn.RNN (i usuń zmienną c0).")
print("Uruchom trening ponownie. Czy zwykłe RNN potrafi rozwiązać ten problem?")
```
