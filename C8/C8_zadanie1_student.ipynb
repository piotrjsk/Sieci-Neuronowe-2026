Przed Tobą prosta konwolucyjna sieć neuronowa. Ma przyjmować na wejściu obrazy ze zbioru MNIST (wymiar: 1 kanał, 28x28 pikseli). Kod jednak **nie działa**. Podczas próby przepuszczenia danych przez sieć (tzw. *forward pass*), interpreter zgłasza błąd związany z niezgodnością wymiarów mnożonych macierzy (`RuntimeError: mat1 and mat2 shapes cannot be multiplied`).

To najczęstszy błąd przy budowaniu sieci CNN od zera – przejście z warstw splotowych do warstw gęstych (`Linear`) wymaga precyzyjnego obliczenia rozmiaru wyjściowego tensora.

**Twoje zadanie:**
1. Uruchom poniższą komórkę i przeczytaj błąd.
2. Wewnątrz metody `forward`, tuż po operacji spłaszczania (`x.view(...)`), dopisz funkcję `print(x.shape)`.
3. Uruchom kod ponownie, zobacz, jaki jest rzeczywisty wymiar tensora, i popraw parametr `in_features` w warstwie `self.fc1`.

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
class ZepsutaSiec(nn.Module):
    def __init__(self):
        super(ZepsutaSiec, self).__init__()
        # Symulujemy wejście dla obrazka MNIST: 1 kanał, 28x28 pikseli
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=16, kernel_size=3, padding=1)
        self.conv2 = nn.Conv2d(in_channels=16, out_channels=32, kernel_size=3, padding=1)
        
        # TUTAJ JEST BŁĄD: in_features ustawione na losową wartość 100
        self.fc1 = nn.Linear(in_features=100, out_features=10)

    def forward(self, x):
        # 1. Warstwa splotowa + ReLU + MaxPool
        x = self.conv1(x)
        x = F.relu(x)
        x = F.max_pool2d(x, kernel_size=2)
        
        # 2. Warstwa splotowa + ReLU + MaxPool
        x = self.conv2(x)
        x = F.relu(x)
        x = F.max_pool2d(x, kernel_size=2)
        
        # 3. Spłaszczanie tensorów do jednego wektora
        x = x.view(x.size(0), -1) 
        
        # >>> MIEJSCE NA TWOJĄ LINIJKĘ KODU: print(x.shape) <<<
        
        # 4. Warstwa w pełni połączona (Linear)
        x = self.fc1(x)
        return x
```

```python
# Tworzymy atrapę obrazka (1 obrazek w batchu, 1 kanał, 28x28 pikseli)
dummy_input = torch.randn(1, 1, 28, 28)
```

```python
# Inicjalizujemy model
model = ZepsutaSiec()
```

```python
# Próbujemy przepuścić dane przez sieć
print("Rozpoczynamy propagację w przód (forward pass)...")
output = model(dummy_input)
print("Sukces! Wynik:", output.shape)
```
